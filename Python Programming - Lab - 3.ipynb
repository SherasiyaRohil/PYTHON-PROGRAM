### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
y=int(input("enter year:"))
if((y%4==0 and not y%100==0) or y%400==0):
    print(f"{y} is leap year")
else:
    print(f"{y} not leap year")
```

07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
n1=int(input("enter num1:"))
n2=int(input("enter num2:"))
choice=int(input("enter your choice 1,2,3,4,5:"))
match choice:
    case 1:
        print("sun:",n1+n2)
    case 2:
        print("sub:",n1-n2)
    case 3:
        print("mul:",n1*n2)
    case 4:
        print("div:",n1/n2)
    case 5:
        print("mod:",n1%n2)
    case _:
        print("invalid choice")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
m1,m2,m3,m4,m5=input("enter a 5 sub mark:").split(',')
m1=int(m1)
m2=int(m2)
m3=int(m3)
m4=int(m4)
m5=int(m5)
per=(m1+m2+m3+m4+m5)*100/500
print(per)

if per<35:
    print("Fail below")
elif per<=45 :
    print("Pass Class ")
elif per<=60 :
    print("second Class ")  
elif per<=70 :
    print("First Class ") 
else:
    print("very good")    
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit=int(input("enter unit :"))

if unit>=1 or unit<=50:
    ans=unit*2.60
elif unit>50 or unit<=100:
    ans=50*2.60+(unit-50)*3.25
elif unit>100 or unit<=200:
    ans=50*2.60+(unit-50)*3.25+(unit-100)*5.26
else:
    ans=50*2.60+(unit-50)*3.25+(unit-100)*5.26+(unit-200)*8.45
print(ans)
    
    
```
