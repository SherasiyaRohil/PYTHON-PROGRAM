### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
T = (4, 7, 9, 11, 15)

result = []

for i in range(len(T)):
    n = T[i]
    
    if n > 1:
        prime = True
        for j in range(2, n):
            if n % j == 0:
                prime = False
                break
        
        if prime:
            result.append((i, n))

print(tuple(result))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
result = []


for i in range(len(T[0])):
    sum = 0
    for x in T:
        sum += x[i]
    result.append(sum)

print(result)

    
````
