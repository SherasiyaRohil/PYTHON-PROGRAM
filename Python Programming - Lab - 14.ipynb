### 03) WAP to generate line chart for given data, but create your custom axis.

```python
x = range(1,11)
y = [1,5,9,7,5,6,3,2,4,9]

ax=plt.axes()
ax.set_xlim([0,50])
ax.set_ylim([-10,10])

ax.set_xticks(range(0,50,5))
ax.set_xticks(range(0,50,5))
ax.grid()

plt.plot(x,y)

plt.show()
```

### 04) WAP to plot a line chart showing temperature variation during a week. Mark the lowest and highest temperatue.

```python
Days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
Temperature = [20, 32, 31, 33, 34, 43, 36]
plt.plot(Days,Temperature,marker="*",color='red',ls=':')
plt.annotate("heighest temp.",xy=["Sat",43])
plt.annotate("lowest temp.",xy=["Mon",20])
plt.show()
```

### 07) WAP to create a horizontal bar chart representing the number of books sold in different categories. 
### The program should also display the number of books sold at the end of the respective bar.

```python
Categories = ["Fiction", "Science", "History", "Technology", "Biography"]
BooksSold = [120, 90, 70, 110, 60]
bars=plt.barh(Categories,BooksSold,color=['r','pink','m','y','k'])
plt.bar_label(bars)
plt.xlabel("Categories")
plt.ylabel("BooksSold ")

plt.show()
```
