### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
c = float(input("Enter temperature in Celsius: "))
f = (c * 9/5) + 32
print("Fahrenheit:", f)

f = float(input("Enter temperature in Fahrenheit: "))
c = (f - 32) * 5/9
print("Celsius:", c)
```
