### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s1='Python3@Lab'
d=0
u=0
l=0
s=0
for i in s1:
    if i.isupper():
        u+=1
    elif i.islower():
        l+=1
    elif i.isdigit():
        d+=1
    else:
        s+=1

print("upper",u)
print("Lower",l)
print("Digit",d)
print("special",s)
    
        
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter string: ")
result = ""

for i in s:
    if i >= 'A' and i <= 'Z':
        result += chr(ord(i) + 32)
    elif i >= 'a' and i <= 'z':
        result += chr(ord(i) - 32)
    else:
        result += i

print(result)
```

### 11) Username Validation

### 15) Custom Split
- `apple,banana,grape`
- `10|20|30`
- `A-B-C-D`

```python
s = input("Enter string: ")
s1 = input("Enter separator: ")

word = ""
result = []

for ch in s:
    if ch == s1:
        result.append(word)
        word = ""
    else:
        word += ch

result.append(word)

print(result)
```
